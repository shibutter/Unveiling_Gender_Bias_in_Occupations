```python
def compare_fightin_scores(df_gpt, df_llama, word):
    """
    This function takes two dataframes containing adjectives or verbs and their Fightin' word scores,
    merges them, and plots a comparison of Fightin' word scores between GPT-3.5 and LLama 2.
    
    Parameters:
    df_gpt (pd.DataFrame): DataFrame containing GPT-3.5 Fightin' word scores.
    df_llama (pd.DataFrame): DataFrame containing LLama 2 Fightin' word scores.
    word (str): The word type to compare, such as 'Adjective' or 'Verb'.
    """
    # Renaming columns for clarity
    df_gpt.columns = [word, 'Fightin_Word_Score_GPT']
    df_llama.columns = [word, 'Fightin_Word_Score_LLAMA']

    # Merging the dataframes on the specified word column
    merged_df = pd.merge(df_gpt, df_llama, on=word, how='inner')

    # Plotting a comparison of Fightin' word scores between GPT-3.5 and LLama 2
    plt.figure(figsize=(14, 7))
    plt.scatter(merged_df['Fightin_Word_Score_GPT'], merged_df['Fightin_Word_Score_LLAMA'], alpha=0.5)
    plt.title('Comparison of Fightin\' Word Scores between GPT-3.5 and LLama 2 on {}'.format(word))
    plt.xlabel('GPT-3.5 Fightin\' Word Score')
    plt.ylabel('LLama 2 Fightin\' Word Score')
    plt.grid(True)
    plt.show()

    return merged_df
```
